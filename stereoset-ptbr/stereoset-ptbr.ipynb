```python
import json
import pandas as pd

with open("results.json", encoding="utf-8") as f:
    results = json.load(f)

rows = []

for model, metrics in results.items():

    overall = metrics["overall"]

    rows.append({
        "Modelo": model,
        "LM": overall["LM Score"],
        "SS": overall["SS Score"],
        "ICAT": overall["ICAT Score"]
    })

df = pd.DataFrame(rows)

df.sort_values(
    "ICAT",
    ascending=False
)
```
